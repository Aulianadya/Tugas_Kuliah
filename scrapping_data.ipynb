```python
import pandas as pd
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# Download resource NLTK jika belum ada
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
# Download the missing 'punkt_tab' resource
nltk.download('punkt_tab')


# Load dataset dari file CSV
filename = "tweets-data/en.csv"  # Sesuaikan dengan lokasi file
try:
    df = pd.read_csv(filename, delimiter=",")  # Pastikan delimiter sesuai
except FileNotFoundError:
    print(f"File {filename} tidak ditemukan.")
    exit()

# Pastikan kolom 'full_text' ada dalam dataset
if 'full_text' not in df.columns:
    print("Kolom 'full_text' tidak ditemukan dalam dataset.")
    exit()

# Hapus duplikasi berdasarkan teks
df.drop_duplicates(subset=["full_text"], keep="first", inplace=True)

# Inisialisasi stopwords dan lemmatizer
stop_words = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()

# Fungsi preprocessing teks
def preprocess_text(text):
    if pd.isna(text):  # Tangani nilai NaN
        return ""

    text = re.sub(r"@\w+", "", text)  # Hilangkan mention
    text = re.sub(r"http\S+|www\S+", "", text)  # Hilangkan URL
    text = re.sub(r"[^a-zA-Z\s]", "", text)  # Hanya simpan huruf dan spasi
    text = text.lower()  # Konversi ke huruf kecil
    words = word_tokenize(text)  # Tokenisasi
    words = [word for word in words if word not in stop_words]  # Hapus stopwords
    words = [lemmatizer.lemmatize(word) for word in words]  # Lematisasi
    return " ".join(words)

# Terapkan preprocessing ke setiap tweet
df["clean_text"] = df["full_text"].apply(preprocess_text)

# Simpan hasil preprocessing ke file baru
df.to_csv("tweets_cleaned.csv", index=False)

print("Preprocessing selesai! Data tersimpan di 'tweets_cleaned.csv'.")
```
